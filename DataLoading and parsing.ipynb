<h2 style = "color:blue">Reading JSON File</h2>

<div style = "background-color:lightblue; color:green">JSON ( Java-Script Object Notation) stores data in key-values format like python. There are mainly three methods to read json file.

```python
df1 = pd.read_json('/home/kartika/Downloads/weather.json')
df1.head(5)
```

<div style = "background-color:lightblue; color:green">pd.read_json helps to directly json file into pandas dataframe.

```python
json_data = [
    {'id':1,'Name':'Alpha'},
    {'id':2, 'Name':'Beta'},
    {'id':3, 'Name':'Gamma'},
    {'id':4, 'Name':'Theta'}
]
content = json.dumps(json_data)
with open('/home/kartika/Downloads/selfcreatedjsonfile.json','w') as f:
    f.write(content)
```

```python
with open('/home/kartika/Downloads/selfcreatedjsonfile.json','r') as f:
    data = f.read()
content = json.loads(data)
```

```python
content
```

<div style = "background-color:lightblue; color:green"> Another method of reading json file is by using json module. json.jumps converts python data (list of dictionary) to json format. json.loads converts json data to python data format.

```python
dff = DataFrame(content)
dff
```

<div style = "background-color:lightblue; color:green">If data stored in json file as dictionary then it can be directly converted to DataFrame using DataFrame function.
