<h2 style = "color:blue">6.Joining numpy arrays.

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])
comb = np.concatenate((arr1,arr2))
print(comb)
arr3 = np.array([[1,2],
                 [3,4]])
arr4 = np.array([[5,6],
                 [7,8]])

print(np.concatenate((arr3,arr4),axis = 0)) #by default
print(np.concatenate((arr3,arr4),axis = 1))
```

<div style = "color:green; background-color:lightblue">Here, we joined two arrays using concatenate method.<br>
When there is axis = 0 , which is a default value, the arrays are joined rows wise. It is like array1 is stacked upon
array2. <br>
When there is axis = 1, the arrays are joined column wise. It is like array2 is placed right of array1 and joined.</div>

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([11,12,13,14,15])
comb = np.stack((arr1,arr2),axis = 0)
print(comb)
print(np.stack((arr1,arr2),axis = 1))
```

<div style = "color:green; background-color:lightblue">Here, we joined two arrays using stack method. The only difference is that
using stack method adds one more dimension to the joined array.<br>First we joined two 1D arrays and we got 2D arrays.
</div>

```python
arr1 = np.array([[1,2],
                 [3,4]])
arr2 = np.array([[5,6],
                 [7,8]])
print(np.stack((arr1,arr2),axis = 2))
```

<div style = "color:green; background-color:lightblue">As using stack give us one more dimension, when we create a combined array
using two 2D arrays, we get a 3D arrays and the way elements are combined can be modified with one more axis that is axis = 2
which join elements as two 2D arrays are stack upon each other. If this seems to be confusing you can take two piece of paper, make
each 2D array one each paper and stack them upon one another and compare the elements you get from the given program.</div>

```python
a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6],
              [7,8]])
print(np.hstack((a,b)))
print(np.vstack((a,b)))
print(np.dstack((a,b)))
```

<div style = "color:green; background-color:lightblue">hstack is the shorcut method for horizontal stack , vstack is the shortcut for

<h2 style="color:blue">1.Broadcasting

<div style ="background-color:lightblue; color:green">Broadcasting is the set of rules NumPy uses to apply arithmetic (and other operations) between arrays of different shapes.

Instead of explicitly copying data to match shapes, NumPy stretches the smaller array virtually (without actually using extra memory) so that element-wise operations can be performed.</div>

```python
arr = np.array([3,4,3])
print(arr*5)
```

<div style ="background-color:lightblue; color:green">Here numpy broadcast 5 into [5,5,5] to match the shape of the array to perform multiplication.
</div>

<div style ="background-color:lightblue; color:green">Rules for broadcasting<br>
i.The size of the dimensions of the array must be same or one of them must be of size 1.<br>
ii. Compare the dimensions from the right and missing dimenion is condidered 1.
<br>Example: (3,) broadcasted to (1,3) and (3,1) are compatible and resulting array is (3,3)</div>

```python
arr1 = np.array([1,2,3])
arr2 = np.array([[1,2,3],
                 [4,5,5],
                 [4,5,6]])
print(arr1+arr2)
```

<div style ="background-color:lightblue; color:green">arr1 broadcasted to [[1,2,3],[1,2,3],[1,2,3]].</div>

<h2 style = "color:blue">2.ufunc

<div style="background-color:lightblue;color:green">ufunc stands for universal function. They are numpy functions that take arrays as arguemtent
and perform element-wise operations faster and efficient than using explicit loops.</div>

```python
arr1 = np.array([1,2,3])
arr2 = np.array([3,4,5])
arr3 = np.array([4,5,3])
np.add(arr1,arr2,arr3)
```

<div style="background-color:lightblue;color:green">Difference between np.add() and np.sum().<br>
Here sum() is a reduction function that aggregates values from single array, whereas add is a universal function that perform element-wise operations between multiple arrays.</div>

```python
arr1 = np.array([1,2,3])
arr2 = np.array([3,4,5])
print(np.subtract(arr1,arr2))
print(np.multiply(arr1,arr2))
print(np.divide(arr1,arr2))
print(np.power(arr1,2))
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([3,4,5])
print(np.add.reduce(arr1))
print(np.multiply.accumulate(arr2))
```

<div style="background-color:lightblue;color:green">Here we have used reduce and accumulate methods to aggregate values as universal functions
operates on two or more arrays. Also, add.reduce() acts and sum() function and add.accumulate acts as cumsum().</div>

<h2 style="color:blue">2.1 Declaring your own ufunc

```python
def my_add(x,y):
    return x+y

addition = np.vectorize(my_add)
arr1 = np.array([1,2,3])
arr2 = np.array([3,4,5])
print(addition(arr1,arr2))
```

```python
def power(x,y):
    return x**y

pow = np.vectorize(power)
arr1 = np.array([1,2,3])
arr2 = np.array([3,4,5])
print(pow(arr1,arr2))
```

<div style="background-color:lightblue;color:green">Here we can use vectorize method to create a python function that work as ufunc that takes
arrays and perform element-wise operation. But it is not actually as fast as normal ufunc.</div>
