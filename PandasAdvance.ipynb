<h1 style="color:blue">Data Aggregation and Group Operation</h1>

```python
import numpy as np
import pandas as pd 

from pandas import Series
from pandas import DataFrame
```

<h1 style = "color:blue">Grouping

<div style="color:green; background-color:lightblue">Data aggregation refers to combining multiple values in single summary values.<br>
Data Aggregation is one of the part of Group Operation.

<div style="color:green; background-color:lightblue">Group operation refers to splitting data into parts, applying function to them and finally combining to get final result.<br>
Group aggregation : Split ---> Apply ----> Combine

```python
df = DataFrame({'key1':list('abcedacd'),
                'key2':[1,2,3,2,1,1,3,2],

                'data1':np.random.randint(1,100,size=8),
                'data2':np.random.randint(1,100,size=8)})
print(df)
```

```python
groups = df.groupby('key1')
for x,y in groups:
    print(x,y)
```

```python
groups.get_group('a')
```

```python
groups.size()
```

```python
groups.count()
```

<div style="color:green; background-color:lightblue">Groupby function can be used to split data into different groups using columns of dataframe.

```python
grouped = df['data1'].groupby(df['key1'])
grouped.mean()
```

```python
grouped = df['data1'].groupby(df['key2'])
grouped.mean()
```

<div style="color:green; background-color:lightblue">This is one of the way by which we can calculate the mean of particular column using particular key label using groupby function.


```python
means = df['data2'].groupby([df['key1'],df['key2']]).mean()
means
```

<div style="color:green; background-color:lightblue">The data can be grouped by passing multiple keys, resulting series with herirarchial indexing.

```python
df.groupby('key1').mean()
```

```python
df.groupby(['key1','key2']).mean()
```

<div style="color:green; background-color:lightblue">In this way we can split entire dataframe into groups and apply function.

```python
df = DataFrame(np.random.randint(1,100,size=(5,5)),columns = list('abcde'))
df
```

```python
mapping = {'a':'FirstSum',
           'b':'FirstSum',
           'c':'SecondSum',
           'd':'FirstSum',
           'e':'SecondSum'}

df.T.groupby(mapping).sum().T
```

```python
series = Series(mapping)
df.T.groupby(series).sum().T
```

<div style="color:green; background-color:lightblue">In this way we can group data using dictionary and Series.
    . Note: Using axis = 1, with groupby function is depreciated in newer version, so transpose is used.

```python
df = DataFrame({'Names':['Alpha','Beta','Theta','Gamma','Delta','Lambda']})
df
```

```python
group = df['Names'].groupby(df['Names'].str.len())
for x,y in group:
    print(x,y)
```

<div style = "background-color:lightblue; color:green">In this way grouping can be done by passing functions as well.</div>

```python
df = DataFrame(np.random.randint(1,100,size=(4,4)),index = [['a','b','a','b'],np.arange(4)])
df
```

```python
df.groupby(level=1).sum()
```

<h2 style = "color:blue">Aggregating</h2>

```python

group = df.groupby('key1')
for x,y in group:
    print(x,y)
```

```python
agg = df.groupby('key1').agg({'data1':['min','max','sum']})
print(agg)
```

<div style = "background-color:lightblue; color:green">After grouping the dataframe using any column we can perform various aggregate functions on specific column in that grouped data.<br>
Using agg allows us to perform more than one aggregate function over grouped data.
</div>

```python
data = {

    "Customer": np.random.choice(["Alice", "Bob", "Charlie", "David", "Eva"], 30),
    "City": np.random.choice(["Kathmandu", "Pokhara", "Lalitpur", "Biratnagar"], 30),
    "Product": np.random.choice(["Laptop", "Phone", "Tablet", "Headphones"], 30),
    "Category": np.random.choice(["Electronics", "Accessories"], 30),
    "Quantity": np.random.randint(1, 5, 30),
    "Price": np.random.randint(100, 1000, 30),
    
}
df = DataFrame(data)
print(df.head())
```

```python
print(df.groupby('City')['Price'].sum())
```

<div style = "background-color:lightblue; color:green">This calculates the total sales per city.</div>

```python
print(df.groupby('Product')['Quantity'].sum())
```

<div style = "background-color:lightblue; color:green">This calculates the total quantity sold per product.

```python
print(df.groupby('City')['Price'].agg(['mean','sum','min','max']))
#or df.groupby('City').agg({'Price':['mean','sum','min','max']})
```

<div style = "background-color:lightblue; color:green">This calculates the sales statistics per city.

```python
print(df.groupby(['City','Product'])['Price'].agg('sum'))
```

<div style = "background-color:lightblue; color:green">This calculates total sales per city as well as per product.

 <h2 style = "color:blue">Transform

```python
print(df.head())
```

```python
df["City_Avg_Revenue"] = df.groupby("City")["Price"].transform("mean")
```

```python
print(df.head())
```

<div style = "background-color:lightblue; color:green">transform() is used when performing aggregate function by grouping but keeping the original dataframe shape.<br>
It is useful when adding new column based group calculations.

<h2 style = "color:blue">Filtering

```python
print(df.head())
```

```python
print(df.groupby('City').filter(lambda x: x['Price'].mean() > 500))
```

<div style = "background-color:lightblue; color:green">filter can be used to filter one or more rows from the grouped data based on the given condition.

<div style = "background-color:lightblue; color:green">Pivot table is way of summarizing large datasets by transforming data from the original dataset by grouping and aggregating.

<h1 style = "color:blue">Time Series</h1>

<div style = "background-color:lightblue; color:green">Time series is simply sequence of observations(data points) indexed by time order.</div>

<div style = "background-color:lightblue; color:green">Types of time series data:<br>
1.Regualr Time series(Fixed Frequency Time series):<br>
Data is recorded at consistent, equally spaced time intervals.<br>
<br>
2.Irregular Time Series:<br>
Data is recorded at inconsistent, random time intervals.
<br>
<br>
3.Time-Stamped:<br>
Each value is tied to each time stamp.
<br>
<br>
4. Period Time Series:<br>
Data represents time span(intervals) instead of specific moment.

```python
from datetime import datetime
```

```python
now = datetime.now()
print(now)
print(f"{now.year},{now.month},{now.day},{now.hour},{now.minute},{now.second}")
```

```python
delta = datetime(2006, 10, 19) - now
print(delta)
```

```python
string_date = "2024-06-01"
date_object = datetime.strptime(string_date, "%Y-%m-%d")
print(date_object)

dates = pd.to_datetime(["2024-01-01", "2025-01-02", "2023-12-31"], format = "%Y-%m-%d")
print(dates)
```

<div style = "background-color:lightblue; color:green">string date can be converted to datetime object using datetime.strptime and pd.to_datetime methods.<br><br>
pd.to_datetime is pandas way and more flexible as it converts list, series or column of string to datetime data.<br><br>
datetime.strptime is python way and converts string to datetime only one as time with strict format specifier.

<h2 style = "color:blue">Periods and Period Arithmetic

<div style = "background-color:lightblue; color:green">Periods denote time span rather than specific moment.<br>
timestamp → 2024-01-01 10:00:00 (exact time)<br>
Period → 2024-01 (entire month)

```python
periods = pd.period_range('2025-01',periods = 5, freq = 'M')
print(periods)
```

<div style = "background-color:lightblue; color:green">In this way periods range can be declared. <br>It is different from date_range : pd.date_range('2024-01',periods = 5, freq = 'M') as internally it is treated as date time, but period is span not specific time. 

```python
periods = periods +1
print(periods)
periods = periods -1
print(periods)
```

<div style = "background-color:lightblue; color:green">Here periods arithmetic is similar to normal arithmetic. Adding or subtracting increase or decrease the span repectively.

```python
period = pd.Period('2025-01', freq = 'M')
period = period.to_timestamp()
print(period)
```

<div style = "background-color:lightblue; color:green">Here, periods can be converted to timestamp using to_timestamp method. You can clearly see the dtype for periods and datetime.

```python
period = period.to_period(freq = 'M')
print(period)
```

<div style = "background-color:lightblue; color:green">Similary, timestamp can be converted to period using to_period method.

<h2 style = "color:blue">Resampaling and Frequency Conversion

<div style = "background-color:lightblue; color:green">Resampaling refers to chainging the frequency of the given timeseries.

<div style = "background-color:lightblue; color:green">Downsampaling refers to converting high frequency to low frequency.<br>
Daily to Monthly<br>
Hourly to Daily

```python
data = Series(np.random.randint(1,100,8), index = pd.date_range('2026-01-01',periods = 8, freq = 'D'))
print(data)
```

```python
print(data.resample('ME').mean())
```

<div style = "background-color:lightblue; color:green">THe given timeseries has frequency of daily date. <br> 
We can resample it to monthly frequency using resample() method. Here, 'ME' stands for month end frequency. 
<br>Since we are resamplaing daily to monthly,  mean() function is used to calculate the average value for each month to avoid fluctuation.

<div style = "background-color:lightblue; color:green">Usamplaing refers to converting low frequency to high frequency.<br>
Daily to Hourly<br>
Montly to Daily

```python
print(data.resample("h").ffill())
```

<div style = "background-color:lightblue; color:green">In this way we can convert low freq(daily) to high freq(hourly) using resample method. <br>
Since, upsampaling adds new timeslots some missing values are introduced which are handaled using fifill, bfill or interploate().


```python
data1 = data.asfreq('D')
print(data1)
```

<div style = "background-color:lightblue; color:green">asfreq simply converts the frequency of given timeseries. <br>
Unlike resample method that first group the given series and then use reduction function to give single value.

<h2 style = "color:blue">Moving Window Function

<div style = "background-color:lightblue; color:green">A moving window takes fixed-sized subset of data and applies function as it slides over time.<br>



```python
data = Series([30,40,50,60,70])
print(data.rolling(window = 3).mean())
```

<div style = "background-color:lightblue; color:green">windows:<br>
First two rows doesn't have enough data for window(3).<br>
[30,40,50]-mean<br>
[40,50,60]-mean<br>
[50,60,70]-mean

<div style = "background-color:lightblue; color:green">In the above example we got NaN in the first two rows, since no enough data for window(3). We can set the minimum number of data for window using min_periods parameter.

```python
print(data.rolling(3, min_periods = 1).mean())
```

```python
time = Series(np.random.randint(1,100,5),index = pd.date_range('2024-01-01', periods = 5, freq = 'D'))
print(time)
print(time.rolling('3D').mean())
```

<div style = "background-color:lightblue; color:green">This is time based rolling. Instead of last three rows it calculates based on last three days.

```python
print(time.ewm(span = 3).mean())
```

<div style = "background-color:lightblue; color:green">ewm(exponentially weighted mean) gives more weight to recent values and ensures smoothness.<br>
Overall windows functions ensure smoothness in noisy data, filling gaps and clear detection.
