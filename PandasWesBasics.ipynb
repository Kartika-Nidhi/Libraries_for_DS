<h2 style="color:blue">6. Arithmetic and Data alignment</h2>

<h3 style="color:blue">6.1 Data alignment in series and dataframes</h2>

```python
s1 = Series([1,2,3,4],index = ['a','b','c','d'])
s2 = Series([4,3,2,5,2,5], index = ['d','b','g','i','c','a'])
print(s1+s2)
```

<div style="background-color:lightblue; color:green">When we perform arithmetic on two objects of different arrangement of index then the resulting object will have the union of indexes of those two objects. The value of only those indexes that are common in both will be obtained and others will be NaN.</div>

```python
df1 = DataFrame(np.arange(9).reshape((3,3)),index = list('acd'),columns = ['One','Three','Five'])
df2 = DataFrame(np.arange(12).reshape((3,4)), index = list('bag'),columns = ['Three','Four','Two','One'])
print(df1)
print("=========================")
print(df2)
print("=========================")
print(df1+df2)
```

```python
df1 = DataFrame(np.arange(9).reshape((3,3)),index = list('acd'),columns = [1,3,5])
df2 = DataFrame(np.arange(12).reshape((3,4)), index = list('bag'),columns = [3,4,2,1])
print(df1)
print("=========================")
print(df2)
print("=========================")
print(df1+df2)
```

<div style="background-color:lightblue; color:green">Alignment works on the basis of data labels. The resulting object will have the indexes as the union of index pairs of two dataframes. After union they are arraned on the basis of alphabetical order and only values of common data labels  are obtained and other remains NaN.</div>

<h3 style="color:blue">6.2 Filling value for NaN</h3>

```python
df1 = DataFrame(np.arange(9).reshape((3,3)),index = list('acd'),columns = [1,3,5])
df2 = DataFrame(np.arange(12).reshape((3,4)), index = list('bag'),columns = [3,4,2,1])
print(df1)
print("=========================")
print(df2)
print("=========================")
print(df1+df2)
```

```python
df3 = df1+df2
# df3[df3.isna()] = 10
print(df3.fillna(10))  
```

<div style="background-color:lightblue; color:green">We can simply assign a specific value to the resulting object with NaN using boolean conditioning and .isna() or fillna() method. There is also another way to perform same operation.</div>

```python
df1 = DataFrame(np.arange(9).reshape((3,3)),index = list('acd'),columns = [1,3,5])
df2 = DataFrame(np.arange(12).reshape((3,4)), index = list('bag'),columns = [3,4,2,1])
print(df1.add(df2,fill_value = 10))
```

<div style="background-color:lightblue; color:green">fill_value argument is passed to add method along with another object. It helps to assign value in the resulting object if atleast one value is not present in one of the object. If the value is not present in both then the resultig object will still have NaN. If you want to completely replace Nan with some value you can use .isna() or fillna() methods as explained above.</div>

<h3 style="color:blue">8.1 Using .map() method

<div style="background-color:lightblue; color:green">Mapping is the process transforming pandas objects by replacing values of series or dataframe columns according defined conditions.</div>

```python
series = Series(np.arange(6),index=list('abcdef'))
```

```python
print(series)
```

```python
mapped_series = series.map(lambda x: x**2)
print(mapped_series)
```

<div style="background-color:lightblue; color:green">.map() methods takes a function as argument and apply it to the series and transform it which is called mapping. Here, map is used to series object. It can take series, functions or dictionary for mapping.</div>

```python
series = Series(['B','C','A','E','D'])
map_dict = {
    'A':'Excellent',
    'B':'VeryGood',
    'C':'Good',
    'D':'Bad',
    'E':'BetterLuckNextTime'
}
mapped_series = series.map(map_dict)
print(mapped_series)
```

<div style="background-color:lightblue; color:green">we can even use dictionary or series for mapping.</div>

```python
df = DataFrame({
    'Gender':['Male','Female','Female','Male','Female'],
    'Experience':[2,4,2,5,4],
    'Salary':[x*1000 for x in range(1,6)]
})

Gender_mapper = {'Male':'M',
                 'Female':'F'}

df['Gender'] = df['Gender'].map(Gender_mapper)
print(df)
```

<h3 style="color:blue">8.2 Using .apply()</h3>

<div style="background-color:lightblue;color:green">.apply() function is used to apply other function along any axis of a dataframe.</div>


```python
df = DataFrame({
    'Gender':['Male','Female','Female','Male','Female'],
    'Experience_Work':[2,4,2,5,4],
    'Experience_Business':[2,3,2,5,2],
    'Salary':[x*1000 for x in range(1,6)]
})
print(df)
```

```python
def expadder(row):
    total_exp = row['Experience_Work']+row['Experience_Business']
    return total_exp

df.apply(expadder,axis = 1)


```

<div style="background-color:lightblue; color:green"> In the above example, we have dataset for persons categorized by their gender and their experience in work and business. First we defined a normal python function that will take inputs from each rows(Experience_Work and Experience_Busiiness) of the dataframe and add them to give overall experience. We used apply function to apply this function to each row of the data frame or along axis = 1. </div>

<h2 style ="color:blue">11. Corelation and Covariance

<h3 style="color:blue">11.1 Covariance</h3>

```python
obj = Series(np.arange(1,6),index = list('abcde'))
print(obj)
```

```python
obj2 = Series(np.arange(1,6), index = list('abcde'))
print(obj2)

obj3 = obj2.apply(lambda x : x-2)
print(obj3)
```

```python
print(obj.cov(obj3))
```

```python
df = DataFrame({'Height':[188,192,155,163],
                'Weight':[60,67,80,59]},
                index = ['Person1','Person2','Person3','Person4'])
print(df)
```

```python
print(df['Height'].cov(df['Weight']))
```

<div style ="background-color:lightblue; color:green"> Covariance tells how two quantity changes together. Positive covariance means both quantity increase/decrease together, whereas negative covariance means one increase and other decrease or vice versa.</div>

<h3 style="color:blue">11.2 Corelation</h3>

<div style ="background-color:lightblue; color:green">Corelation measures how strongly and in which direction two variables are related. It is a standarilized form of covariance and is bounded to -1 to +1 depending upon the strength of relation.
</div>

```python
df = DataFrame({'Height':[188,192,155,163],
                'Weight':[60,67,80,59]},
                index = ['Person1','Person2','Person3','Person4'])
print(df)
```

```python
print(df['Height'].corr(df['Weight']))
      
```
