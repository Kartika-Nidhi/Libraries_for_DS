<h3 style = "color:blue">Discretization and Binning
</h3>

<div style="color:green; background-color:lightblue">Discrete data can only take specific, seperate values.<br>
It usually takes integer value.<br>
It is usually counted.<br>
For example: Number of students per class = [40,33,25]

<div style="color:green; background-color:lightblue">Continuous data can take any values within a range.<br>
It can take decimals, fractions.<br>
It is usually measured.<br>
For example: Height of Plants = [3.4,5.6,4,4]

<div style="color:green; background-color:lightblue">Binning is the process of grouping continuous numeric data into discrete categories for analysis.

```python
ages = Series([12,75,13,59,34,23,75,23,33,83,23])

bins = [18,25,40,60,80]
age_category = pd.cut(ages,bins,right = False)  # By default intervals are (], left is exclusive and right is inclusive
age_category
```

<div style="color:green; background-color:lightblue">pd.cut can be used to group continous data into discrete category.

```python
age_category.value_counts()
```

```python
group_names = ['Young','Adult','Middle Aged','Senior']
pd.cut(ages, bins, labels = group_names)
```

<div style="color:green; background-color:lightblue">The numeric intervals can be labeled as various ways by passing list of new labels in labels parameter.

```python
pd.cut(ages,4,precision = 2) # precision = 2 limits the decimal value upto two digits
```

<div style="color:green; background-color:lightblue">If any integer is passed , pandas automatically divides the whole data into equal width bins. It is calculated by using maximum and minimum value from the data giving us less control compared to explicitly defining and passing bin width.

```python
q=pd.qcut(ages,4, precision = 2)
q.value_counts()
```

<div style="color:green; background-color:lightblue">pd.cut() divides the data into equal width bins, whereas pd.qcut() divides data into equal size bins.

<h3 style = "color:blue">Regular Expression (regex)</h3>

```python
import re
```

<div style="color:green; background-color:lightblue">Regular Expression (regex) is a pattern used to search, match, extract or replace complex string patterns in text.

<div style="color:green; background-color:lightblue">patters are normally written as raw string that treats \ as literal character rather than an escape character.

```python
csv = pd.read_csv('/home/kartika/Downloads/people_data.csv')
csv
```

```python
text = '''Shelby	Terrell	Male	elijah57@example.net	1945-10-26	Games developer
	Phillip	Summers	Female	bethany14@example.com	NaN	Phytotherapist
	Kristine	Travis	NaN	bthompson@example.com	1992-07-02	NaN
	Yesenia	Martinez	Male	kaitlinkaiser@example.com	2017-08-03	Market researcher
	Lori	Todd	Male	NaN	NaN	Veterinary surgeon'''
```

```python
pattern = r"\d{4}-\d{2}-\d{2}"
regex = re.compile(pattern)

# matches = re.findall(pattern,text)
#matches
print(regex.findall(text))
print(regex.search(text))
print(regex.match(text))
```

<div style="color:green; background-color:lightblue">First we can create a pattern using raw string and regex tokens that can be compiled by re module and use to find pattern in text. .findall function can be used to search all the patterns matching regex. search only returns the first occurence of pattern and match returns if pattern appears at the beginning of the string.

<div style="color:green; background-color:lightblue">It is recommended to use re.compile to create regex object of the pattern to be used multiple time as it improves readibility and performance.<br>
As the pattern/regex is compiled first and then only other re functions are used. So when same regex is being used multiple time, it is better to create object of it using re.compile.

```python
email_pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"
regex = re.compile(email_pattern, flags = re.IGNORECASE)

regex.findall(text)
```

<div style="color:green; background-color:lightblue">Suppose we want to extract name, domain name from the email. We can seperate each component , () can be use for each pattern.

```python
regex = re.compile(r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})", flags = re.IGNORECASE)
regex.findall(text)
```

<h3 style = "color:blue">Using pd.merge</h3>

```python
df1 = DataFrame({'key':list('aca'),
                 'values1':np.arange(3)})
df2 = DataFrame({'key':list('baccb'),
                 'values2':np.arange(8,13)})
pd.merge(df1,df2, on = 'key')
                
```

<div style="color:green; background-color:lightblue">pd.merge joins datasets based on key columns (common column).<br>
on parameter specify the key column.<br> The resulting merged dataframe has common as well as repeated common rows.

```python
pd.merge(df1,df2, how = 'right')
```

```python
pd.merge(df1,df2, how = 'left')  
```

<div style="color:green; background-color:lightblue"> how parameter takes two arguements namely right(all rows from right dataframe) and left(all rows from left dataframe) that specifies the type of join. Missing value while merging is represented by Nan.

```python
pd.merge(df1,df2, how = 'outer', indicator = True) # defualt inner
```

<div style="color:green; background-color:lightblue">Similarly, how parameter takes another two arguments namely inner(common rows only) and outer(rows from both dataframes).

```python
df3 = DataFrame({'A':np.arange(5),'B':np.arange(5,10)})
df4 = DataFrame({'A':np.arange(4,9),'B':np.arange(15,20)})
pd.merge(df3,df4, left_index = True, right_index = True, suffixes = ('First','Second'))
```

<div style="color:green; background-color:lightblue">In this way all the common columns in the dataframes can merged in single dataframe.

<h3 style = "color:blue">Pivoting Long to Wide Format and Vice Versa</h3>

<div style="color:green; background-color:lightblue">In Wide Format, each variable has its seperate column and data is spread horizontally.<br>
In Long Format, one column stores variables and another stores values, and each row represent observation.

```python
df = DataFrame({'product':['Table','Desk','Chair'],
                'width':[4,2,6],
                'height':[5,6,3],
                'length':[5,3,7]})
df
               
```

```python
Long_format = df.melt(id_vars = ['product'],var_name = 'Variables',value_name = 'Value')
Long_format
```

<div style="color:green; background-color:lightblue">Wide format data can be converted to long format using pd.melt function.<br>
It has three parameters: id_vars: column to be fixed, var_name : column for variable and value_name: column for values.

```python
wide_format = Long_format.pivot(index= ['product'],columns = 'Variables',values = 'Value')
wide_format
```

<div style="color:green; background-color:lightblue">Long format data can be converted to wide format using pivot function.
