<h1 style = "color:blue">Bar Graph</h1>

```python
#x and y axis values
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)

#labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph', fontsize = 10)


#color, size and plotting
color = ['cyan', 'magenta', 'yellow', 'blue', 'green']
plt.bar(ax,ay, color = color, edgecolor = 'blue', linewidth = 2, linestyle = ":")   #bars[1].set_color('red') to change color of a single bar
plt.figuresize=(20,5)
plt.legend(['Bar 1', 'Bar 2', 'Bar 3', 'Bar 4', 'Bar 5'])
plt.show()
```
