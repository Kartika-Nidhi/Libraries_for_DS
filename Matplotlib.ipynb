<h1 style = "color:green">Matplotlib</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("/home/kartika/Datasets/products.csv")

df.head(20)
# df.info()
```

<div style = "color:blue; font-size:20px">Bar Graph

```python
plt.bar(df['category'].unique(), df['cocoa_percent'].groupby(df['category']).mean())

plt.title("Cocoa Percent of Products by Category")
plt.xlabel("Category")
plt.ylabel("Cocoa Percent")
plt.show()
```

<div style = "background-color:lightblue; color:green">The above bar graph shows the average cocoa percent of chocolates by their category.
