<h1 style = "color:blue">1. Bar Graph</h1>

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: Compare values across different categories</li>
<li>Use when: Data is categorical</li>
</ul>

```python
#x and y axis values
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)

#labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph', fontsize = 10)


#color, size and plotting
color = ['cyan', 'magenta', 'yellow', 'blue', 'green']
plt.bar(ax,ay, color = color, edgecolor = 'blue', linewidth = 2, linestyle = ":", label = ax)   #bars[1].set_color('red') to change color of a single bar
plt.figure(figsize =(10,4))
plt.show()
```

```python
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)
az= np.random.randint(1,100,5)

plt.title("Bar Graph")
plt.xlabel("Names", fontsize = 15)
plt.ylabel("Freq", fontsize = 15)

plt.bar(ax,ay, color = 'green', edgecolor = 'lightgreen', linewidth = 2, linestyle = ":", label = 'Freq1') 
plt.bar(ax,az, color = 'blue', edgecolor = 'lightblue', linewidth = 2, linestyle = ":", label = 'Freq2')
plt.legend()
plt.show()  
```

```python
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)
az= np.random.randint(1,100,5)

plt.title("Bar Graph")
plt.xlabel("Names", fontsize = 15)
plt.ylabel("Freq", fontsize = 15)



plt.barh(ax,ay, color = 'green', edgecolor = 'lightgreen', linewidth = 2, linestyle = ":", label = 'Freq1') 
plt.barh(ax,az, color = 'blue', edgecolor = 'lightblue', linewidth = 2, linestyle = ":", label = 'Freq2')
plt.legend()
plt.show()  
```

<div style = "background-color:lightblue; color:green">We can create horizontal bar graph using barh.

```python
data = DataFrame({
    'Freq1':ay,
    'Freq2':az
},index = ['Alpha','Beta','Gamma','Delta','Lambda'])




data.plot(kind = 'bar')
plt.xticks(rotation = 20)
plt.show()
```

<div style = "background-color:lightblue; color:green">Dataframe can be used to plot the bars of different dataseries with same x lable, side by side, instead of top and bottom.

```python
x = np.arange(10)
y = np.arange(10)
plt.bar( x ,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.axhline(y = 2, color = 'red', linestyle = '--')
plt.axvline(x = 5, color = 'blue', linestyle = '--')

for i in range(len(x)):
    plt.text(x[i], y[i], str(y[i]), ha = 'center', va = 'bottom')

plt.show()
```

<div style = "background-color:lightblue; color:green">In this way we can add vertical and horizontal line in bar graph.

<h1 style = "color:blue">2. Scatter Plot

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: Show relationship or correlation or dispersion between two variables</li>
<li>Use when: To detect patterns, clusters, or outliers</li>
</ul>

```python
x = np.random.randint(1,50,10)
y1 = np.random.randint(1,50,10)
y2 = np.random.randint(1,50,10)

plt.title('Scatter Plot')
plt.xlabel('X-axis', fontsize = 20)
plt.ylabel('Y-axis', fontsize = 20)


plt.scatter(x,y1, c= y1, edgecolor = 'blue', linewidth = 2, s = 400, label = 'Data Points1', alpha = 0.5, marker = ".", cmap ='plasma')
plt.scatter(x,y2, c= y2, edgecolor = 'green', linewidth = 2, s = 400, label = 'Data Points2', alpha = 0.5, marker = ".", cmap ='viridis')

cbar1 = plt.colorbar()
cbar2 = plt.colorbar()

cbar1.set_label("Colorbar Based on Height")
cbar2.set_label("Colorbar Based on Height")
plt.legend(loc = 'upper right')
plt.show()

```

<h1 style = "color:blue">4. Piechart

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: Show proportion or percentage contribution of parts to a whole</li>
<li>Use when: Comparing relative sizes</li>
</ul>

```python
values1 = np.random.randint(1,100,5)
values2 = np.random.randint(1,100,5)
labels = ['Alpha','Beta','Gamma','Delta','Lambda']
plt.title("Pie Chart", fontsize = 20, color = 'lightblue')

explode = [0.2,0.0,0.0,0.0,0.0]  #to explode a slice
color1 = ['cyan', 'magenta', 'yellow', 'blue', 'green']






plt.pie(values1, labels=labels, explode=explode, colors=color1, autopct="%1.1f%%", shadow = True, radius = 1, labeldistance = 1.1, 
        startangle = 90,counterclock = False, 
        wedgeprops = {'edgecolor': 'blue', 'linewidth': 2, 'linestyle': ':'},
        textprops = {'fontsize': 10, 'color': 'black'})


# plt.legend(loc = 2, fontsize = 10)
plt.figure(figsize=(4,10))
plt.show()
```

<h1 style = "color:blue">5. Line Plot

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: Show trends or changes over a continous variable (usually time)</li>
<li>Use when: Data is continous and order matters</li>
</ul>


```python
ay1 = np.random.randint(1,500,5)
ay2 = np.random.randint(1,500,5)
ax= pd.period_range(start = '2024', periods = 5, freq = 'Y').strftime('%Y')

```

```python
plt.plot(ax,ay1, label = 'Line1', color = 'g', linestyle = "--", linewidth = 2, marker = 'o', markersize = 4, markerfacecolor = 'red',
             markeredgecolor = 'blue', markeredgewidth = 2)

plt.plot(ax,ay2, label = 'Line2', color = 'b', linestyle = "--", linewidth = 2, marker = 'o', markersize = 4, markerfacecolor = 'cyan',
             markeredgecolor = 'blue', markeredgewidth = 2)

plt.title("Line Graph", fontsize = 20, color = 'lightblue')
plt.legend()

plt.grid(True)
plt.xlabel("Dates")
plt.ylabel("Frequency")
plt.show()
```

<h1 style = "color:blue">8. Stack Plot

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: Show compostion of categories along with total, over continuous period</li>
<li>Use when: comparing parts within groups, data is continuous(usually time sereis)</li>
</ul>

```python
x = ['Jan','Feb','Mar','Apr','May']
sales1= np.random.randint(1,100,len(x))
sales2 = np.random.randint(1,100,len(x))
sales3 = np.random.randint(1,100,len(x))
print(x,sales1, sales2, sales3)




plt.stackplot(x, sales1, sales2, sales3, labels =['Sales 1','Sales 2','Sales 3'], colors = ['r','g','b'])
plt.yticks(np.arange(0,200,10))
plt.grid(True)

plt.legend(loc = 'upper right')
plt.show()
```

<div style = "background-color:lightblue; color:green">Here, the top boundary shows the total sales from Jan to May. The coloured area shows contributin of each sales type in total from Jan to May.

<h1 style = "color:blue">9. Step Plot

<div style = "background-color:lightblue; color:green">
<ul>
<li>Purpose: To show changes in values that occur at specific points, where the value stays contant between changes and suddenly jumps.</li>
<li>Use when: Data that remains constant and suddenly change over time.</li>
</ul>

```python
x = np.arange(4)
y = np.arange(4)

plt.step(x,y, color = 'magenta', marker ="o", ms = 5, mfc = 'g')

plt.title("Step Plot")
plt.grid(True)
plt.show()
```

<h1 style = "color:blue">10. Filling between plots

```python
x = np.random.randint(1,100,5)
y = np.random.randint(1,100,5)

plt.plot(x,y, color = 'g', marker = 'o', markersize = 8, markerfacecolor ='r')


# plt.fill_between(x = [20,40], y1 =20, y2 =60, color = 'lightblue', alpha = 0.5, label = 'Area between y1 and y2')
plt.fill_between(x,y, color = 'lightblue', where = ((x>20) & (x<60)))

plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

<h1 style="color:blue">11. Subplots

<h2 style = "color:blue">11.1 Using plt.subplot()

```python
x = ['Alpha','Beta','Gamma','Delta','Sigma']
y = np.random.randint(1,100,5)

plt.subplot(1,2,1) #(rows, columns, index)[(1)|(2)]
plt.bar(x,y, color = plt.cm.Greens(np.linspace(0.2,0.8,5)))
plt.title('Bar Graph')

plt.subplot(1,2,2) #(rows, columns, index)[(1)|(2)]
plt.pie(y, colors = ['red','blue','green','magenta','cyan'], labels = x, autopct = "%1.1f%%", shadow = True, startangle = 90)
plt.title('Pie Plot')


plt.show()


```

<h2 style = "color:blue">11.2 Using plt.subplots()


```python
x = ['Alpha','Beta','Gamma','Delta','Sigma']
y = np.random.randint(1,100,5)

fig, ax = plt.subplots(1,2) #(rows, columns, index)[(1)|(2)]


ax[0].bar(x,y, color = plt.cm.Greens(np.linspace(0.2,0.8,5)))
ax[0].set_title("Bar plot")


ax[1].pie(y, colors = ['red','blue','green','magenta','cyan'], labels = x, autopct = "%1.1f%%", shadow = True, startangle = 90)
ax[1].set_title('Pie plot')

plt.subplots_adjust(wspace = 0.2)
plt.tight_layout()
plt.show()
```
