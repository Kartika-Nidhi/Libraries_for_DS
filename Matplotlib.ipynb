<h1 style = "color:blue">Bar Graph</h1>

```python
#x and y axis values
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)

#labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph', fontsize = 10)


#color, size and plotting
color = ['cyan', 'magenta', 'yellow', 'blue', 'green']
plt.bar(ax,ay, color = color, edgecolor = 'blue', linewidth = 2, linestyle = ":", label = ax)   #bars[1].set_color('red') to change color of a single bar
plt.figuresize=(20,5)
plt.legend()
plt.show()
```

```python
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)
az= np.random.randint(1,100,5)

plt.title("Bar Graph")
plt.xlabel("Names", fontsize = 15)
plt.ylabel("Freq", fontsize = 15)

plt.bar(ax,ay, color = 'green', edgecolor = 'lightgreen', linewidth = 2, linestyle = ":", label = 'Freq1') 
plt.bar(ax,az, color = 'blue', edgecolor = 'lightblue', linewidth = 2, linestyle = ":", label = 'Freq2')
plt.legend()
plt.show()  
```

```python
ax = ['Alpha','Beta','Gamma','Delta','Lambda']
ay = np.random.randint(1,100,5)
az= np.random.randint(1,100,5)

plt.title("Bar Graph")
plt.xlabel("Names", fontsize = 15)
plt.ylabel("Freq", fontsize = 15)



plt.barh(ax,ay, color = 'green', edgecolor = 'lightgreen', linewidth = 2, linestyle = ":", label = 'Freq1') 
plt.barh(ax,az, color = 'blue', edgecolor = 'lightblue', linewidth = 2, linestyle = ":", label = 'Freq2')
plt.legend()
plt.show()  
```

<div style = "background-color:lightblue; color:green">We can create horizontal bar graph using barh.

```python
data = DataFrame({
    'Freq1':ay,
    'Freq2':az
},index = ['Alpha','Beta','Gamma','Delta','Lambda'])




data.plot(kind = 'bar')
plt.xticks(rotation = 20)
plt.show()
```

<div style = "background-color:lightblue; color:green">Dataframe can be used to plot the bars of different dataseries with same x lable, side by side, instead of top and bottom.

```python
plt.bar(np.arange(10),np.arange(10))
plt.xlabel = 'X-axis'
plt.ylabel = 'Y-axis'

plt.axhline(y = 2, color = 'red', linestyle = '--')
plt.axvline(x = 5, color = 'blue', linestyle = '--')

plt.show()
```

<div style = "background-color:lightblue; color:green">In this way we can add vertical and horizontal line in bar graph.
